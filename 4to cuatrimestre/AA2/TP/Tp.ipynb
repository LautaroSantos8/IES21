```python
# Selecciono las columnas necesarias
df_colectivos=df_colectivos[['id','geometry.coordinates','properties.ultima_posicion_time']]
df_colectivos.head()
```

```python
df_colectivos.info()
```

```python
# Transformo mi columna de coordenadas
df = pd.DataFrame()
for x in df_colectivos['geometry.coordinates']:
  string = x
  characters = "[]"
  string = ''.join( x for x in string if x not in characters)
  df=df.append({'coordenada':string},ignore_index=True)
df.head()
```

```python
df_colectivos
```

```python
# Separo mi coordenada en Latitud y Longitud, ademas de ordenar los datos de las fechas
df_colectivos=pd.concat([df_colectivos,df],axis=1)
df_colectivos[['Latitud','Longitud']]=df_colectivos['coordenada'].str.split(',',expand=True)
import re 
df= pd.DataFrame()
#df_colectivos['properties.ultima_posicion_time'].str.split(('T,-'),expand=True)
for x in df_colectivos['properties.ultima_posicion_time']:
    res = re.split('T|-', x)
    df=df.append({'fecha':res[3]},ignore_index=True)
df_colectivos= pd.concat([df_colectivos,df],axis=1)
df_colectivos.drop(['geometry.coordinates','coordenada','properties.ultima_posicion_time'],axis=1,inplace=True)
df_colectivos.head()
```

```python
df_colectivos
```

```python
# Selecciono el tipo de datos de cada columna
df_colectivos['Latitud']=df_colectivos['Latitud'].astype(float)
df_colectivos['Longitud']=df_colectivos['Longitud'].astype(float)
df_colectivos['fecha']=pd.to_datetime(df_colectivos['fecha'])
```

```python
# Elijo un solo colectivo para predecir su comportamiento
colectivo_fijado=df_colectivos.loc[df_colectivos['id']==69]
```

```python
colectivo_fijado 
```

```python
# Elimino los datos duplicados
colectivo_fijado=colectivo_fijado.drop_duplicates()
len(colectivo_fijado)
```

## DataSet Datos

```python
colectivo_fijado
```

```python
# Genero una combinacion de coordenadas y fechas, para así calcular el tiempo que tardará en desplazarse entre coordenadas
cord_inicial_L=[]
cord_inicial_La=[]
tiempo_inicial=[]
cord_final_L=[]
cord_final_La=[]
tiempo_final=[]

for fn in  range(colectivo_fijado.iloc[1:].shape[0]):
  for ini in range(colectivo_fijado.iloc[0:fn].shape[0]):
     cord_inicial_L.append(colectivo_fijado.iloc[ini]["Longitud"])
     tiempo_inicial.append(colectivo_fijado.iloc[ini]["fecha"])
     cord_final_L.append(colectivo_fijado.iloc[fn]["Longitud"])
     cord_inicial_La.append(colectivo_fijado.iloc[ini]["Latitud"])
     tiempo_final.append(colectivo_fijado.iloc[fn]["fecha"])
     cord_final_La.append(colectivo_fijado.iloc[fn]["Latitud"])
```

```python
# Creo un dataframe
df_coordenadas= pd.DataFrame()
```

```python
df_coordenadas['cord_inicial_L']=cord_inicial_L
df_coordenadas['cord_inicial_La']=cord_inicial_La
df_coordenadas['Tiempo_inicial']=tiempo_inicial
df_coordenadas['cord_final_L']=cord_final_L
df_coordenadas['cord_final_La']=cord_final_La
df_coordenadas['Tiempo_final']=tiempo_final
```

```python
#lo  pasamos  a segundos y elimino mis columnas de fechas
df_coordenadas["target"] = (df_coordenadas['Tiempo_final'] - df_coordenadas['Tiempo_inicial']).dt.total_seconds()
df_coordenadas.drop(['Tiempo_inicial','Tiempo_final'],axis=1,inplace=True)
```

```python
df_coordenadas
```

```python
from sklearn.model_selection import train_test_split
x= df_coordenadas.drop('target',axis=1)
y= df_coordenadas['target']
# Separo mis datos en un set de datos conocidos, usads para el entrenamiento, y un set de datos desconocidos que simulan datos nuevos 
# para cuando se ponga en funcionamiento el modelo.
x_conocido, x_desconocido, y_conocido, y_desconocido= train_test_split(x,y,test_size=0.1, random_state=123)
# Separo mi set de entrenamiento y testeo 
x_train, x_test, y_train, y_test = train_test_split(x_conocido,y_conocido,test_size=0.2, random_state=123)
```

```python
df_coordenadas.info()
```

### Modelos 

```python
# Librerias
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression  
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

#### Regresion lineal

```python
RegLin=LinearRegression()
parameters = {}
gsearchRL= GridSearchCV(RegLin,param_grid=parameters, cv=5,scoring='neg_root_mean_squared_error')
gsearchRL.fit(x_train,y_train)
```

```python
(gsearchRL.best_score_)*-1
```

#### Ridge 

```python
ridge=Ridge(random_state=123)
parameters_ridge = {'alpha':[1e-07,1e-06,1e-05]}
gsearch_ridge= GridSearchCV(ridge,param_grid=parameters_ridge, cv=5,scoring='neg_root_mean_squared_error')
gsearch_ridge.fit(x_train,y_train)
```

```python
gsearch_ridge.best_params_
```

```python
gsearch_ridge.best_score_*-1
```

#### Lasso

```python
lasso=Lasso(random_state=123)
parameters_lasso = {'alpha':[1e-11,1e-10,1e-09]}
gsearch_lasso= GridSearchCV(lasso,param_grid=parameters_lasso, cv=5,scoring='neg_root_mean_squared_error')
gsearch_lasso.fit(x_train,y_train)
```

```python
gsearch_lasso.best_params_
```

```python
gsearch_lasso.best_score_*-1
```

#### Arbol

```python
arbol=DecisionTreeRegressor(random_state=123)
parameters_arbol = {'splitter':('best','random'),'max_depth':[14,15,16]}
gsearch_arbol= GridSearchCV(arbol,param_grid=parameters_arbol, cv=5,scoring='neg_root_mean_squared_error')
gsearch_arbol.fit(x_train,y_train)
```

```python
gsearch_arbol.best_params_
```

```python
gsearch_arbol.best_score_*-1
```

#### Comparacion

```python
print('RMSE Regresion Lineal: ',gsearchRL.cv_results_['mean_test_score'][gsearchRL.best_index_]*-1, '+/-' , 2*gsearchRL.cv_results_['std_test_score'][gsearchRL.best_index_] , '(95%)')
print('RMSE Ridge: ',gsearch_ridge.cv_results_['mean_test_score'][gsearch_ridge.best_index_]*-1, '+/-' , 2*gsearch_ridge.cv_results_['std_test_score'][gsearch_ridge.best_index_] , '(95%)')
print('RMSE Lasso: ',gsearch_lasso.cv_results_['mean_test_score'][gsearch_lasso.best_index_]*-1, '+/-' , 2*gsearch_lasso.cv_results_['std_test_score'][gsearch_lasso.best_index_] , '(95%)')
print('RMSE Arbol: ',gsearch_arbol.cv_results_['mean_test_score'][gsearch_arbol.best_index_]*-1, '+/-' , 2*gsearch_arbol.cv_results_['std_test_score'][gsearch_arbol.best_index_] , '(95%)')
```
