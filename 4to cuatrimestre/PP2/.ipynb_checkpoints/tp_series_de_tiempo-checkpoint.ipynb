## DataSet para entrenar

```python
df_final=pd.merge(df_train,df_oil,on='date',how='left')
# Completo los datos nulos 
df_final=df_final.fillna(method='bfill')
df_final=pd.merge(df_final,df_holidays_events,on='date',how='left')
df_final=df_final.merge(df_store,on='store_nbr')
# Completo datos nulos
df_final.loc[(df_final['type_x'].isna()),'type_x'] = 'Common Day'
# Elimino columnas que no necesito
df_final.drop(['locale','locale_name','description','transferred','city','state'],axis=1,inplace=True)
df_final
```

```python
# Pasaje de todas las columnas a tipo numerico
df_final["type_x"] = df_final["type_x"].map({"Holiday": 1, "Common Day": 2})
df_final["type_y"] = df_final["type_y"].map({"A": 1, "B": 2, "C": 3, "D": 4, "E": 5,})
df_final.replace(list(df_final['family'].unique()), list(np.arange(0,len(list(df_final['family'].unique())))),inplace=True)
```

```python
df_final=df_final.set_index('date')
df_final.head()
```

### Funciones

#### Funcion para grid_search_forecaster

```python
def modelos(df,forecaster,param_grid, lags_grid):
    df=df.resample('D').mean()
    df=df.fillna(0)   
    x=df.drop('sales',axis=1)
    y= df['sales']
    
    steps=15  
  # Valores de x
    datos_train_x= x[:-steps]
    datos_test_x= x[-steps:]
  # Valores de y
    datos_train_y= y[:-steps]
    datos_test_y= y[-steps:]  

    resultados_grid = grid_search_forecaster(forecaster        = forecaster,
                                            y                  = datos_train_y,
                                            exog               = datos_train_x,
                                            param_grid         = param_grid,
                                            lags_grid          = lags_grid,
                                            steps              = steps,
                                            refit              = False,
                                            metric             = 'mean_squared_error',
                                            initial_train_size = int(len(datos_train_x)*0.5),
                                            fixed_train_size   = False,
                                            return_best        = True,
                                            verbose            = False)
    return resultados_grid
```

#### Funcion para grafico

```python
def grafico(datos_test_y, predicciones):
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=datos_test_y.index, y=datos_test_y, name='Test', line=dict(color='firebrick', width=2)))
    fig.add_trace(go.Scatter(x=predicciones.index, y=predicciones, name='prediccion', line=dict(color='royalblue', width=2)))
    grafico= fig
    return grafico 
```

### Modelo

```python
# Fijo mi valor de steps
steps=15
# Genero un for anidado para entrenar los 25 modelos necesarios
result={}
for store in df_final['store_nbr'].unique():
  result[store]={}
  for family in df_final['family'].unique():
    result[store][family]={}
    # Establesco el set de datos para trabajar, dependiendo el numero de store y la familia
    result[store][family]['df']= df_final.loc[(df_final['store_nbr']==store) & 
                                              (df_final['family']==family)].drop(['store_nbr','family','id'],axis=1)
    # Al Set de datos le completo el indice, pasando dia a dia, completando los posibles dias faltantes
    result[store][family]['df']=result[store][family]['df'].resample('D').mean()
    # Completo los datos nulos con 0
    result[store][family]['df']=result[store][family]['df'].fillna(0)
    
    # Genero mi modelo forecaster, por terminos de velocidad, usaré Lasso
    result[store][family]['forecaster']= ForecasterAutoreg(
                                                          regressor = Lasso(random_state=123),
                                                          lags      = 12)
    # Parametros de Lasso
    result[store][family]['param_grid']= {'alpha': [1e-6,1e-5,1e-4,0.001,0.01]}
    # Cantidad de Lags
    result[store][family]['lags_grid']= [30,40,50,60,70,80]
    
    #Determino x e y de mi Set de datos
    result[store][family]['x']=result[store][family]['df'].drop('sales',axis=1)
    result[store][family]['y']=result[store][family]['df']['sales']

    # Separo mi train y test
    result[store][family]['datos_train_x']= result[store][family]['x'][:-steps]
    result[store][family]['datos_test_x']= result[store][family]['x'][-steps:]

    result[store][family]['datos_train_y']= result[store][family]['y'][:-steps]
    result[store][family]['datos_test_y']= result[store][family]['y'][-steps:]

    # Entreno los modelos con los datos de mi Set de datos completo, mi forecaster, los parametros y los lags
    result[store][family]['modelos']= modelos(result[store][family]['df'], result[store][family]['forecaster'],
                                              result[store][family]['param_grid'],result[store][family]['lags_grid'])
    
    # Predigo los valores 
    result[store][family]['predicciones']= result[store][family]['forecaster'].predict(steps=15,
                                                                                       exog=result[store][family]['datos_test_x'])
    # Calculo el error
    result[store][family]['error'] = mean_squared_error(y_true=result[store][family]['datos_test_y'], 
                                                                 y_pred = result[store][family]['predicciones'], squared=False)
    
    # Para mayor entendimiento, genero una grafica 
    result[store][family]['grafico'] = grafico(result[store][family]['datos_test_y'], result[store][family]['predicciones'])
```
