# REEMPLAZA LOS NULOS DE LA COLUMNAS DIRECTOR Y COUNTRY CON EL STRING INCOMPLETO

```python
df['director'].replace(np.nan, 'INCOMPLETO', inplace=True)
df['country'].replace(np.nan, 'INCOMPLETO', inplace=True)
```

# ELIMINA LAS RESTANTES COLUMNAS QUE TENGAN VALORES NULOS

```python
df.dropna(axis=1, inplace=True)
```

# TRANSFORMAR LA COLUMNA TYPE CATEGORICA EN NUMERICA Y ELIMINAR LA COLUMNA type_TV Show

```python
df = pd.get_dummies(df, columns=['type'])
```

```python
df.drop(['type_TV Show'],axis=1, inplace=True)
```

# ELIMINAR COLUMNAS listed_in Y description

```python
df.drop(['listed_in', 'description'],axis=1, inplace=True)
```

# CAMBIAR LOS NOMBRES DE LAS COLUMNAS A ESPAÑOL

```python
titulos=['Id', 'Título', 'Director', 'País', 'Año', 'Duración', 'Película']
df.columns=titulos
```

# CONVIERTA A MAYUSCULA Y QUITE LOS ESPACIOS DEL COMIENZO Y FINAL DE LA COLUMNA Título

```python
df.Título = df["Título"].str.upper()
df.Título = df["Título"].str.strip()
```

# CONTAR PELICULAS Y SERIES, EXPORTAR DATOS A UN ARCHIVO FORMATO JSON

```python
df.query('Película==0').count()
```
