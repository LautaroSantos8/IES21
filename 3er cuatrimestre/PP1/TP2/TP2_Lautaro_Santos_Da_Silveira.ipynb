### Empiezo a corregir las unidades 

```python
#Genero una copia del Set de datos original para así poder modificarlo a mi gusto
dfcorregido=df.copy()
#Corrijo los datos, cambiando las unidades de medida si es necesario
dfcorregido['Altura'].loc[dfcorregido['Altura'] >= 3.0] = (dfcorregido['Altura']/100)
dfcorregido['Peso'].loc[dfcorregido['Peso'] >= 600] = (dfcorregido['Peso']/1000)
dfcorregido['PA_max'].loc[dfcorregido['PA_max'] >= 33] = (dfcorregido['PA_max']/10)
#Reemplazo los valores nulos cargados como 0 a np.nan
dfcorregido=dfcorregido.replace(0,np.nan)
dfcorregido
```

### Genero la Columna de IMC

```python
#IMC es igual a peso/altura^2
dfcorregido['IMC']= (dfcorregido['Peso']/(dfcorregido['Altura']**2)).round(2)
dfcorregido
```

```python
df_Corr= dfcorregido.corr()
df_Corr.round(2)
```

De la matriz de correlacion podemos sacar las siguientes conclusiones:

*   Peso: Relacion directa con Altura, sin relacion con edad, poca con Presion maxima y Relacion directa con IMC
*   Edad: sin relacion con Altura, sin relacion con Peso, Relacion directa con Presion maxima, sin relacion con IMC
*   PA_max: sin relacion con Altura, sin relacion con Peso, Relacion directa con Edad, sin relacion con IMC
*   IMC: sin relacion con Altura, Relacion directa con Peso, sin relacion con Edad, Sin relacion con Presion maxima



### Graficas

**Proporcion de Hombres y Mujeres en el estudio**

```python
# Creo variables dummy (0,1) para la columna sexo ya que es categorica
sexo_dummy = pd.get_dummies(dfcorregido['Sexo'], prefix='sexo')
# Para sacar la cantidad total de hombres y mujeres, sumo todos los 1 de cada columna y lo guardo en variables para luego generar un diccionario con estas
x=sexo_dummy['sexo_F'].sum()
y= sexo_dummy['sexo_M'].sum()
datos= {
    'Genero' : ['Mujeres','Hombres'],
    'Cantidad' : [x,y]
}
# Genero un Dataframe con los datos del diccionario anterior
Df_genero= pd.DataFrame(datos)
#importo la libreria plotly para luego generar el grafico de pie
import plotly.express as px
fig = px.pie(Df_genero, values='Cantidad',names='Genero',title='Proporcion de Hombres y Mujeres en el estudio')
fig.show()
```

En la proporcion de hombres y mujeres, se puede ver que la proporcion de cada es parecida, por lo que los datos no quedarian sesgados si esta variable tuviese gran importancia

**Proporcion de edad en el Estudio**

```python
#Genero separaciones de la edad con respecto al grupo (Juventud, adultez y vejez) para luego cargar los datos en un diccionario. En el diccionario
# debo especificar la columna, ya que cada variale es un DataFrame
Juventud=dfcorregido.loc[dfcorregido['Edad']<=29]
Edad_Adulta=dfcorregido.loc[(dfcorregido['Edad']<=59) & (dfcorregido['Edad']>=30)]
Vejez=dfcorregido.loc[dfcorregido['Edad']>=60]
datos= {
    'Etapas' : ['Juventud','Adultez','Vejez'],
    'Cantidad' : [Juventud['Edad'].count(),Edad_Adulta['Edad'].count(),Vejez['Edad'].count()]
}
#Cargo los datos del diccionario en un Dataframe vacio
Df_Edad= pd.DataFrame(datos)
#importo la libreria plotly para luego generar el grafico de pie
import plotly.express as px
fig = px.pie(Df_Edad, values='Cantidad',names='Etapas',title='Proporcion de edades en el estudio')
fig.show()
```

Del grafico anterior se puede observar que la mayoria de datos cargados son adultos, seguido por ancianos y la menor proporcion son jovenes 

**Analisis de IMC**

```python
#Genero separaciones del IMC con respecto al grupo (Delgado, Normal, Sobrepeso, Obesidad, Obesidad morvida) para luego cargar los datos en un diccionario.
# En el diccionario debo especificar la columna, ya que cada variale es un DataFrame
Delgado=dfcorregido.loc[dfcorregido['IMC']<18.5]
Normal=dfcorregido.loc[(dfcorregido['IMC']>=18.5) & (dfcorregido['IMC']<=24.9)]
Sobrepeso=dfcorregido.loc[(dfcorregido['IMC']>=25) & (dfcorregido['IMC']<=29.9)]
Obesidad = dfcorregido.loc[(dfcorregido['IMC']>=30) & (dfcorregido['IMC']<=39.9)]
Obesidad_morvida = dfcorregido.loc[dfcorregido['IMC']>=40]
datos= {
    'Estado' : ['Delgado','Normal','Sobrepeso','Obesidad','Obesidad Morvida'],
    'Cantidad' : [Delgado['IMC'].count(),Normal['IMC'].count(),Sobrepeso['IMC'].count(),Obesidad['IMC'].count(),Obesidad_morvida['IMC'].count()]
}
#Cargo los datos del diccionario en un Dataframe vacio
Df_Edad= pd.DataFrame(datos)
#importo la libreria plotly para luego generar el grafico de pie
import plotly.express as px
fig = px.pie(Df_Edad, values='Cantidad',names='Estado',title='Estado de IMC en el estudio')
fig.show()
```

Del grafico anterior se puede concluir que en el analisis, el mayor porcentaje de los datos tienen el IMC normal, seguido muy de cerca por Sobrepeso, en menor proporcion se encuentra la Obesidad y seguido de cerca la Obesidad Morvida, el grupo Delgado es muy bajo, siendo solo 49 observaciones

**Proporcion de tipos de Presion Arterial maxima**

```python
#Genero separaciones de PA_max con respecto al grupo (Optima, Normal, Alta, Hipertension grado 1, Hipertension grado 2, Hipertension grado 3) 
#para luego cargar los datos en un diccionario. En el diccionario debo especificar la columna, ya que cada variale es un DataFrame
Optima=dfcorregido.loc[dfcorregido['PA_max']<12]
Normal=dfcorregido.loc[(dfcorregido['PA_max']>=12) & (dfcorregido['PA_max']<13)]
Alta=dfcorregido.loc[(dfcorregido['PA_max']>=13) & (dfcorregido['PA_max']<=13.9)]
Hipertension_grado_1 = dfcorregido.loc[(dfcorregido['PA_max']>=14) & (dfcorregido['PA_max']<15.9)]
Hipertension_grado_2 = dfcorregido.loc[(dfcorregido['PA_max']>=16) & (dfcorregido['PA_max']<18)]
Hipertension_grado_3=dfcorregido.loc[dfcorregido['PA_max']>18]
datos= {
    'Categoría' : ['Optima','Normal','Alta','Hipertension_grado_1','Hipertension_grado_2', 'Hipertension_grado_3'],
    'Cantidad' : [Optima['PA_max'].count(),Normal['PA_max'].count(),Alta['PA_max'].count(),Hipertension_grado_1['PA_max'].count(),Hipertension_grado_2['PA_max'].count(),Hipertension_grado_3['PA_max'].count()]
}
#Cargo los datos del diccionario en un Dataframe vacio
Df_Edad= pd.DataFrame(datos)
#importo la libreria plotly para luego generar el grafico de pie
import plotly.express as px
fig = px.pie(Df_Edad, values='Cantidad',names='Categoría',title='Estado de PA_max')
fig.show()
```

**Relacion entre la edad y la Presion Arterial Maxima:**
*   Se Puede ver que a medida que la persona envejece, la Presion Arterial Maxima aumenta
