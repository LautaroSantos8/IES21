```python
import pandas as pd
import numpy as np
```

```python
Df_Origin = pd.read_csv('/content/drive/MyDrive/Inteligencia Artificial/segundo anÌƒo/primer semestre/Practica Profecionalizante 1/tp/TP3/HNP_StatsData.csv')
Df_Origin=Df_Origin.drop(['Unnamed: 65'],axis=1)
Df_Origin
```

```python
Df_Filtro=Df_Origin[Df_Origin['Indicator Code'].isin(['SP.DYN.CBRT.IN','SP.DYN.LE00.IN','HD.HCI.OVRL','NY.GNP.PCAP.CD'])]
Df_Filtro
```

```python
Df_Corto=pd.melt(Df_Filtro,id_vars=list(Df_Filtro.columns.values[0:4]) , var_name='AÃ±o',value_name='Cantidad')
Df_Corto.sort_values(list(Df_Filtro.columns.values[0:2]))
Df_Corto.round(3)
```

```python
Df_Corto.columns = ['Country_Name', 'Country_Code', 'Indicator_Name', 'Indicator_Code','AÃ±o', 'Cantidad']
```

```python
df_paises = pd.read_csv('/content/drive/MyDrive/Inteligencia Artificial/segundo anÌƒo/primer semestre/Practica Profecionalizante 1/tp/TP3/HNP_StatsCountry.csv')
df_paises
df_paises.drop(['Unnamed: 30'],axis=1)
ingresos = df_paises.loc[df_paises['Income Group']=='Upper middle income']
paises = [y for y in ingresos['Country Code']]
```

```python
Df_PaisSelec= Df_Corto[Df_Corto.Country_Code.isin(paises)]
X=Df_PaisSelec.loc[(Df_PaisSelec['Country_Code']== 'ARG')&(Df_PaisSelec['Indicator_Code']== 'NY.GNP.PCAP.CD')]
X
```

```python
GNI=Df_PaisSelec.loc[Df_PaisSelec['Indicator_Code'] == 'NY.GNP.PCAP.CD']
prom=pd.DataFrame(GNI.groupby('Country_Code')['Cantidad'].agg(np.mean)).round(2)
prom['Country_Code']= prom.index 
import plotly.express as px
fig = px.choropleth(prom, locations='Country_Code',
                    color='Cantidad',
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

```python
natalidad=Df_Corto.loc[Df_Corto['Indicator_Code'] == 'SP.DYN.CBRT.IN']
natalidad.dropna()
prom=pd.DataFrame(natalidad.groupby('Country_Code')['Cantidad'].agg(np.mean))
import plotly.express as px
fig = px.bar(prom, x=prom.index, y='Cantidad')
fig.show()
```
