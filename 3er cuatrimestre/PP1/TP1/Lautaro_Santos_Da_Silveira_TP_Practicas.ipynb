### Tasa de crecimiento poblacional anual media para todos los paÃ­ses en el intervalo 2000-2020

```python
#obtengo los datos del periodo 2000-2020
list1= range(2000,2021)
dfanio= dfPaises[dfPaises.Time.isin(list1)] 
dfanio= dfanio.loc[dfanio['Variant'] =='Medium']
dfanio
```

```python
dfanio["x"]=dfanio.groupby('Location')['PopTotal'].shift(1)
dfanio["crecimiento poblacional"]= (dfanio['PopTotal']/dfanio['x'])
```

```python
dfanio.head(25)
```

10 paises con  mayor tasa de crecimiento poblacional anual entre 2000-2020 ignorando los paises con datos faltantes

```python
dfSinan= dfanio.loc[(dfanio['PopMale']!=0)&(dfanio['PopFemale']!=0)]
```

```python
promedioMax=pd.DataFrame(dfSinan.groupby('Location')['crecimiento poblacional'].agg(np.mean))
promedioMax.sort_values(by=['crecimiento poblacional'], ascending=False).head(10)
```

10 paises con  menor tasa de crecimiento poblacional anual entre 2000-2020

```python
promedioMin=pd.DataFrame(dfSinan.groupby('Location')['crecimiento poblacional'].agg(np.mean))
promedioMin.sort_values(by=['crecimiento poblacional']).head(10)
```

Poblacion Argentina entre 2000/2020

```python
dfanioArg=dfanio.loc[dfanio['Location']=='Argentina']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(15, 6))
sns.set_color_codes("pastel")
sns.barplot(x=dfanioArg['Time'], y=dfanioArg['PopTotal'], data=dfanioArg,label="Mujeres", color="b")
```

Poblacion Argentina entre 2000/2020 separado entre hombres y mujeres

```python
import seaborn as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(15, 6))
sns.set_color_codes("pastel")
sns.barplot(x=dfanioArg['Time'], y=dfanioArg['PopTotal'], data=dfanioArg,label="Mujeres", color="b")
sns.set_color_codes("muted")
sns.barplot(x=dfanioArg['Time'], y=dfanioArg['PopMale'], data=dfanioArg,label="Hombres", color="b")
ax.legend(ncol=1, loc="lower right", frameon=True)
ax.set(xlim=(0, 24), ylabel="",
       xlabel="Poblacion Argentina entre 2000/2020")
sns.despine(left=True, bottom=True)
```
