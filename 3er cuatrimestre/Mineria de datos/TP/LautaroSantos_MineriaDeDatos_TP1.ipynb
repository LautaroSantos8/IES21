```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/Inteligencia Artificial/segundo año/primer semestre/mineria/mineria de datos/TP/users.csv')
df
```

```python
df.isnull().sum()
```

```python
Marcas= ['Ford', 'Chevrolet', 'Honda', 'Mazda','Dodge']
filtro= df[df['car_make'].isin(Marcas)]
grupo = filtro.groupby('car_make')
Grafico=(grupo.agg(np.size))
Grafico['car_make']=Grafico.index
from matplotlib import pyplot as plt
fig = plt.figure(figsize =(10, 7))
plt.bar(Grafico['car_make'],Grafico['id'])
plt.ylabel('Cantidad de usuarios')
plt.xlabel('Marcas')
plt.title('Cantidad de Usuarios por Marca de Vehiculo')
```
