```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([7,25,31,40,48])
y = np.array([83,10,50,75,99])
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y,'o')
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(x,y,marker = 'P')
plt.show()
```

```python
plt.plot(x,y,'o:r')
plt.show()
```

```python
 plt.plot(y, marker = 'o', ms = 10, color= 'magenta',linewidth = '5')
plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.grid()
plt.title('Matplot')
plt.xlabel("AÑo")
plt.ylabel("Cantidad")
plt.show()
```
